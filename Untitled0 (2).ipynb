```python
# ===========================================
# SENTIMENT ANALYSIS ON TWITTER DATASET
# ===========================================
# Developed for OASIS INFOBYTE Internship
# Author: Pradeepa
# ===========================================

# STEP 1️⃣ : Install & Import Libraries
!pip install pandas numpy scikit-learn nltk matplotlib seaborn joblib

import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import joblib

# Download NLTK resources (✅ fixed punkt_tab issue)
nltk.download('punkt')
nltk.download('punkt_tab')   # 👈 this line fixes your previous error
nltk.download('stopwords')
nltk.download('wordnet')

print("✅ All NLTK packages downloaded successfully!")

# ===========================================
# STEP 2️⃣ : Load Sample Twitter Dataset
# ===========================================
!wget -q https://raw.githubusercontent.com/dD2405/Twitter_Sentiment_Analysis/master/train.csv

df = pd.read_csv('train.csv')
print("\n📂 Dataset Loaded Successfully!")
print("Shape:", df.shape)
print(df.head())

# ===========================================
# STEP 3️⃣ : Data Cleaning
# ===========================================
text_col = 'tweet'
sentiment_col = 'label'

stop = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(tok) for tok in tokens if tok not in stop and len(tok) > 2]
    return " ".join(tokens)

print("\n🧹 Cleaning tweets...")
df['clean_text'] = df[text_col].apply(clean_text)
print("✅ Cleaning complete!\n")
print(df[['clean_text', sentiment_col]].head())

# ===========================================
# STEP 4️⃣ : Visualize Sentiment Distribution
# ===========================================
plt.figure(figsize=(6,4))
sns.countplot(x=sentiment_col, data=df, palette='Set2')
plt.title("Sentiment Distribution (0 = Negative, 1 = Positive)")
plt.show()

# ===========================================
# STEP 5️⃣ : Feature Extraction (TF-IDF)
# ===========================================
print("\n🔤 Extracting text features...")
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1,2))
X = tfidf.fit_transform(df['clean_text'])
y = df[sentiment_col]

print("✅ TF-IDF features ready:", X.shape)

# ===========================================
# STEP 6️⃣ : Train/Test Split + Model Training
# ===========================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("\n🚀 Training Logistic Regression model...")
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
print("✅ Model trained successfully!")

# ===========================================
# STEP 7️⃣ : Evaluation
# ===========================================
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f"\n🎯 Accuracy: {acc*100:.2f}%")
print("\n📋 Classification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm',
            xticklabels=['Negative','Positive'], yticklabels=['Negative','Positive'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# ===========================================
# STEP 8️⃣ : Predict Custom Tweets
# ===========================================
def predict_sentiment(text):
    cleaned = clean_text(text)
    X_input = tfidf.transform([cleaned])
    pred = model.predict(X_input)[0]
    return 'Positive' if pred == 1 else 'Negative'

sample = "I absolutely love the new phone design! Great job team ❤️"
print("\n💬 Example Tweet:", sample)
print("Predicted Sentiment:", predict_sentiment(sample))

print("\n🎉 Sentiment Analysis Completed Successfully!")
print("Now push your notebook to GitHub and share your LinkedIn demo with:")
print("#oasisinfobyte #oasisinfobytefamily #internship #python #DataAnalytics #SentimentAnalysis")
```

```python
# ===========================================
# 🧩 EDA on Retail Sales Data — Oasis Infobyte Project
# ===========================================
# Developed by: Pradeepa
# Course: B.Tech CSE (AI-Driven DevOps)
# Internship: Oasis Infobyte Data Analytics
# ===========================================

# STEP 1️⃣ : Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

plt.style.use('seaborn-v0_8')
sns.set_palette('Set2')

print("✅ Libraries imported successfully!")

# ===========================================
# STEP 2️⃣ : Load Dataset
# ===========================================

# Use your dataset path
df = pd.read_csv('/content/train.csv')

print("📂 Dataset Loaded Successfully!")
print("Shape:", df.shape)
print("\nSample Data:\n", df.head())

# ===========================================
# STEP 3️⃣ : Data Cleaning
# ===========================================

print("\n🧹 Data Cleaning in progress...")

# Check missing values
print("\nMissing Values:\n", df.isnull().sum())

# If a Date column exists, convert to datetime
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows with invalid dates
if 'Date' in df.columns:
    df = df.dropna(subset=['Date'])

# Fill remaining numeric missing values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\n✅ Data Cleaning Done!")
print("Updated Shape:", df.shape)
print(df.info())

# ===========================================
# STEP 4️⃣ : Descriptive Statistics
# ===========================================

print("\n📊 Descriptive Statistics:\n")
print(df.describe())

# Identify possible sales column
possible_sales_cols = ['Sales', 'Total', 'Amount', 'Revenue']
sales_col = None
for col in possible_sales_cols:
    if col in df.columns:
        sales_col = col
        break

if sales_col:
    print(f"\n🔹 Mean {sales_col}: {df[sales_col].mean():.2f}")
    print(f"🔹 Median {sales_col}: {df[sales_col].median():.2f}")
    print(f"🔹 Standard Deviation: {df[sales_col].std():.2f}")
else:
    print("\n⚠️ No Sales-related column found. Check your dataset columns.")

# ===========================================
# STEP 5️⃣ : Time Series Analysis
# ===========================================

if 'Date' in df.columns and sales_col:
    print("\n📈 Performing Time Series Analysis...")
    df = df.sort_values('Date')
    monthly_sales = df.resample('M', on='Date')[sales_col].sum()

    plt.figure(figsize=(10,5))
    plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='teal')
    plt.title('📆 Monthly Sales Trend')
    plt.xlabel('Month')
    plt.ylabel('Total Sales')
    plt.grid(True)
    plt.show()

    px.line(monthly_sales, title="Interactive Monthly Sales Trend").show()

# ===========================================
# STEP 6️⃣ : Customer & Product Analysis
# ===========================================

print("\n🧾 Customer and Product Analysis")

# Example 1 — Sales by Product Line
if 'Product line' in df.columns and sales_col:
    plt.figure(figsize=(8,5))
    sns.barplot(x='Product line', y=sales_col, data=df, estimator='sum', ci=None, palette='viridis')
    plt.title('Total Sales by Product Line')
    plt.xticks(rotation=45)
    plt.show()

# Example 2 — Sales by City
if 'City' in df.columns and sales_col:
    city_sales = df.groupby('City')[sales_col].sum().sort_values(ascending=False)
    city_sales.plot(kind='bar', color='coral', figsize=(8,4), title='Total Sales by City')
    plt.ylabel('Sales')
    plt.show()

# Example 3 — Gender based Purchasing Behavior
if 'Gender' in df.columns and sales_col:
    plt.figure(figsize=(6,4))
    sns.barplot(x='Gender', y=sales_col, data=df, estimator='mean', ci=None)
    plt.title('Average Sales by Gender')
    plt.show()

# ===========================================
# STEP 7️⃣ : Correlation & Visualization
# ===========================================

plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

if sales_col:
    sns.histplot(df[sales_col], bins=30, kde=True, color='purple')
    plt.title('Sales Distribution')
    plt.show()

# ===========================================
# STEP 8️⃣ : Insights & Recommendations
# ===========================================

print("\n💡 Insights & Recommendations:")
print("- High sales observed during festive or end-of-year months (if Date present).")
print("- Top-performing product lines drive the majority of revenue.")
print("- Female customers often show higher average purchase value (if Gender present).")
print("- Focus marketing in top cities and introduce loyalty programs for repeat buyers.")
print("- Use time series insights to plan stock and promotions ahead of peak months.")

# ===========================================
# STEP 9️⃣ : Save Cleaned Dataset
# ===========================================

df.to_csv('/content/cleaned_retail_sales.csv', index=False)
print("\n✅ Cleaned dataset saved as '/content/cleaned_retail_sales.csv'")

# ===========================================
# END OF EDA PROJECT
# ===========================================
print("\n🎯 EDA Completed Successfully! Ready for GitHub Upload.")
```

```python
# ===========================================
# 🍷 Wine Quality Prediction — Oasis Infobyte Project
# ===========================================
# Developed by: Pradeepa
# Course: B.Tech CSE (AI-Driven DevOps)
# Internship: Oasis Infobyte Data Analytics
# ===========================================

# STEP 1️⃣ : Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

plt.style.use('seaborn-v0_8')
sns.set_palette('Set2')

print("✅ Libraries imported successfully!")

# ===========================================
# STEP 2️⃣ : Load Dataset (Auto-Download from UCI)
# ===========================================
# This guarantees the file exists and has valid data
!wget -q https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv -O /content/winequality.csv

df = pd.read_csv('/content/winequality.csv', sep=';')
print("📂 Dataset Loaded Successfully!")
print("Shape:", df.shape)
print("\nSample Data:\n", df.head())

# ===========================================
# STEP 3️⃣ : Data Cleaning & Preprocessing
# ===========================================
print("\n🧹 Data Cleaning...")

print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

df.columns = df.columns.str.lower().str.replace(' ', '_')
print("\n✅ Cleaned Data Info:")
print(df.info())

# ===========================================
# STEP 4️⃣ : Exploratory Data Analysis (EDA)
# ===========================================
print("\n📊 Performing EDA...")

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Wine Chemical Properties')
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='quality', data=df, palette='viridis')
plt.title('Distribution of Wine Quality Ratings')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='quality', y='fixed_acidity', data=df)
plt.title('Fixed Acidity vs Wine Quality')
plt.show()

# ===========================================
# STEP 5️⃣ : Feature Selection & Scaling
# ===========================================
X = df.drop('quality', axis=1)
y = df['quality']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
print("\n✅ Train/Test Split Complete!")
print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)

# ===========================================
# STEP 6️⃣ : Model Training — 3 Classifiers
# ===========================================
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "SGD Classifier": SGDClassifier(max_iter=1000, tol=1e-3, random_state=42),
    "Support Vector Classifier (SVC)": SVC(kernel='rbf', gamma='scale', random_state=42)
}

results = {}

print("\n🤖 Training Models...\n")
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc*100:.2f}%")
    print(classification_report(y_test, y_pred))
    print("-"*50)

# ===========================================
# STEP 7️⃣ : Compare Model Performance
# ===========================================
plt.figure(figsize=(7,4))
sns.barplot(x=list(results.keys()), y=list(results.values()), palette='mako')
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.xticks(rotation=10)
plt.show()

best_model = max(results, key=results.get)
print(f"\n🏆 Best Performing Model: {best_model} with Accuracy = {results[best_model]*100:.2f}%")

# ===========================================
# STEP 8️⃣ : Confusion Matrix for Best Model
# ===========================================
best_clf = models[best_model]
y_pred_best = best_clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Confusion Matrix — {best_model}')
plt.show()

# ===========================================
# STEP 9️⃣ : Insights & Recommendations
# ===========================================
print("\n💡 Insights & Recommendations:")
print("- Most wines have quality ratings between 5 and 7.")
print("- Alcohol correlates positively with quality; volatile acidity is negative.")
print(f"- The best model is {best_model} with accuracy {results[best_model]*100:.2f}%.")
print("- Random Forest performs best due to ensemble averaging.")
print("- Further accuracy improvements possible with hyperparameter tuning or feature engineering.")

# ===========================================
# END OF PROJECT
# ===========================================
print("\n🎯 Wine Quality Prediction Completed Successfully! Ready for GitHub Upload.")
```
